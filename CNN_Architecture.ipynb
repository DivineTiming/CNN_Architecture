**Question 1:**
What is the role of filters and feature maps in Convolutional Neural
Network (CNN)?

**Answer 1**
Filters in CNNs act as small windows that scan input images to detect specific patterns like edges or textures by performing convolution operations.
Feature maps are the outputs of these convolutions, representing the detected features across the image and serving as inputs for subsequent layers.

In a Convolutional Neural Network (CNN):

Filters (or Kernels): These are small matrices of numbers that slide over the input image (or feature map from a previous layer). Each filter is designed to detect a specific type of feature, such as edges (horizontal, vertical, diagonal), corners, or textures. When a filter slides over the input, it performs an element-wise multiplication with the portion of the input it covers and then sums the results, producing a single output value. Different filters learn to detect different patterns. The values within the filter (the weights) are learned during the training process.

Feature Maps (or Activation Maps): These are the outputs generated by applying a filter to the input. Each feature map represents the detected presence of a specific feature at different spatial locations in the input. For example, if a filter is designed to detect vertical edges, its corresponding feature map will highlight where vertical edges are present in the input. The higher the value at a particular point in the feature map, the stronger the presence of that feature at that location in the input. These feature maps then serve as inputs to subsequent layers in the CNN, allowing the network to learn increasingly complex and abstract features.

**Question 2**
Explain the concepts of padding and stride in CNNs(Convolutional Neural
Network). How do they affect the output dimensions of feature maps?

**Answer 2**
Padding adds extra pixels around the input image borders to preserve size or control output dimensions, with "same" padding maintaining input size and "valid" reducing it. Stride determines the step size for filter movement; larger strides downsample the output, shrinking feature map dimensions via the formula: output size = (input size - kernel size + 2*padding) / stride + 1.

Padding involves adding extra pixels (usually zeros) around the border of the input image or feature map before applying the convolution operation.

Preserve Spatial Dimensions: Without padding, each convolution operation reduces the spatial dimensions of the feature map. If you perform many layers of convolution without padding, the feature maps can shrink too quickly, potentially losing valuable information from the edges of the input.
Retain Edge Information: Pixels at the edges of an image are only processed once (or very few times) by a filter, whereas pixels in the center are processed multiple times. Padding ensures that edge pixels contribute more to the output, as they are included in more convolution operations.

Different types of Padding:

Valid (No Padding): No padding is applied. The output feature map will be smaller than the input.
Same (Zero-Padding): Padding is added such that the output feature map has the same spatial dimensions as the input, assuming a stride of 1.

Stride refers to the number of pixels the filter shifts over the input image or feature map. A stride of 1 means the filter moves one pixel at a time. A stride of 2 means it moves two pixels at a time, effectively skipping some pixels.

Stride is used for:

Downsampling: Stride is primarily used to reduce the spatial dimensions (width and height) of the feature maps. A larger stride results in a smaller output feature map.
Computational Efficiency: By reducing the size of feature maps, stride helps decrease the computational cost of subsequent layers.

Padding increases the output dimensions (or helps maintain them).
Stride decreases the output dimensions. A larger stride leads to a more significant reduction in size.
These two parameters are crucial for controlling the spatial dimensions throughout the CNN architecture, allowing developers to design networks with appropriate depth and feature map sizes.

**Question 3**
Define receptive field in the context of CNNs. Why is it important for deep architectures?

**Answer 3**
The receptive field is the region of the input image influencing a neuron's activation in a CNN, starting small in early layers for local features like edges and expanding in deeper layers. It is crucial for deep architectures to capture global context, enabling recognition of complex objects beyond local details.

Convolutional Neural Networks (CNNs), the receptive field refers to the region in the input space that a particular feature in a higher layer is looking at or is influenced by. Think of it as the area of the input image that a neuron in a subsequent layer 'sees' when computing its output.

Here's why it's important for deep architectures:

Hierarchical Feature Extraction: As you go deeper into a CNN, each successive layer has a larger receptive field than the previous one. This allows deeper layers to learn more complex and abstract features by combining information from broader regions of the input. For example, early layers might detect edges or textures (small receptive fields), while deeper layers might detect entire objects or parts of objects (large receptive fields) by combining these simpler features.

Contextual Understanding: A larger receptive field means that a neuron can incorporate more contextual information from the input. This is crucial for understanding relationships between different parts of an image and for tasks like object recognition, where the context around an object can be vital.

Efficiency and Parameter Sharing: While the receptive field grows, the number of parameters at each convolutional layer does not necessarily increase proportionally. This is due to weight sharing, a fundamental property of CNNs. Convolutional filters are applied across the entire input, effectively reusing the same set of weights to detect patterns everywhere. This allows deep CNNs to learn rich representations without an explosion of parameters, making them computationally efficient.

Handling Scale Variations: By having varying receptive field sizes across layers, deep CNNs can implicitly handle objects at different scales. Smaller objects might be detected by features with smaller receptive fields in earlier layers, while larger objects require larger receptive fields in deeper layers.

In essence, the progressive increase in receptive field size through a deep CNN is what enables it to build a hierarchical understanding of visual data, moving from local, fine-grained features to global, semantic concepts.

**Question 4**
Discuss how filter size and stride influence the number of parameters in a CNN.

**Answer 4**
Larger filter sizes increase parameters per filter (e.g., 5x5 has 25 weights vs. 3x3's 9), raising total parameters for a given number of filters. Stride affects parameters indirectly by altering feature map sizes, influencing parameters in subsequent layers but not the current conv layer's count.

The filter size (or kernel size) and stride are two crucial hyper-parameters in a Convolutional Neural Network (CNN) that influence its architecture and, consequently, the number of parameters.

Their impact is listed as below:

 - Filter Size (Kernel Size)
The filter size refers to the dimensions of the sliding window (kernel) that convolves over the input. For a single convolutional layer, the number of parameters is calculated as:

Number of Parameters = (Filter Height * Filter Width * Number of Input Channels + 1 (for bias)) * Number of Filters

Direct Impact: As you can see from the formula, the filter height and filter width directly contribute to the number of parameters. A larger filter size (e.g., 5x5 instead of 3x3) will significantly increase the number of parameters for each filter. This is because each filter needs to learn more weights to cover a larger region of the input.

Example: A 3x3 filter operating on 64 input channels has (3*3*64 + 1) parameters per filter. A 5x5 filter on the same input channels has (5*5*64 + 1) parameters per filter. If you have 128 such filters, the total parameters would be much higher for the 5x5 case.
Pros of Larger Filters: Can capture larger spatial patterns and have a wider receptive field in early layers.

Cons of Larger Filters: More parameters, higher computational cost, and potential for overfitting if not enough data.

 - Stride
Stride defines how many pixels the convolution filter shifts across the input image or feature map at each step. While filter size directly impacts the parameters within a convolutional layer, stride's influence on the total number of parameters in a CNN is more indirect, primarily by affecting the size of the output feature maps.

No Direct Impact on Layer Parameters: Stride does not directly change the number of weights or biases within a given convolutional filter. The parameters for a filter remain the same regardless of whether it moves one pixel at a time (stride=1) or two pixels at a time (stride=2).

Indirect Impact on Subsequent Layers: Stride primarily affects the spatial dimensions (height and width) of the output feature map. A larger stride (e.g., stride=2) reduces the size of the output feature map more aggressively than a smaller stride (e.g., stride=1). This reduction in feature map size can then influence the number of parameters in subsequent layers, particularly:

Following Convolutional Layers: If a subsequent convolutional layer takes the output of the current layer as its input, a smaller input feature map due to a larger stride in the preceding layer might lead to fewer computations, but the number of parameters for the subsequent convolutional filter itself still depends on its own filter size, number of input channels (which is the number of filters from the previous layer), and number of its own filters.
Fully Connected Layers: If the CNN architecture flattens the output of a convolutional block into a fully connected layer, a smaller feature map (due to larger strides in earlier layers) means fewer neurons in the flattened input to the fully connected layer. This will result in a significantly reduced number of parameters in the fully connected layers.
Example: A convolutional layer with stride=2 will produce an output feature map roughly half the height and half the width of one with stride=1 (assuming padding is handled similarly). If this output is then fed into a fully connected layer, the number of input connections (and thus parameters) for that fully connected layer will be much smaller for the stride=2 case.

In summary, filter size directly increases the parameters of the convolutional layer itself, as each weight within the filter contributes to the total. Stride does not affect the parameters of the current convolutional layer but indirectly influences the total parameters of the CNN by altering the spatial dimensions of feature maps, which then impacts the parameter count of subsequent layers, especially fully connected layers.

**Question 5**
Compare and contrast different CNN-based architectures like LeNet,
AlexNet, and VGG in terms of depth, filter sizes, and performance.

**Answer 5**
LeNet-5 (1998)
Depth: Relatively shallow, typically 7 layers (2 convolutional, 2 pooling, 3 fully connected).

Filter Sizes: Started with larger filters (e.g., 5x5) and relatively small feature maps.

Key Features: Pioneering work in CNNs, introduced concepts like convolution, pooling, and activation functions. Used for handwritten digit recognition.

Performance: Excellent for its time on tasks like MNIST, but limited by computational power and dataset size for more complex image recognition.
AlexNet (2012)
Depth: Deeper than LeNet, typically 8 layers (5 convolutional, 3 fully connected).

Filter Sizes: Used larger filters in early layers (e.g., 11x11, 5x5) and then smaller 3x3 filters. Incorporated max-pooling.

Key Features: Broke records in the ImageNet Large Scale Visual Recognition Challenge (ILSVRC) 2012. Key innovations included ReLU activation functions (addressing vanishing gradients), dropout (reducing overfitting), and data augmentation. Utilized GPUs for training.
Performance: Significantly outperformed traditional computer vision methods and LeNet on large-scale datasets like ImageNet, demonstrating the power of deep learning for image recognition.

VGG (Visual Geometry Group) (2014)
Depth: Much deeper than AlexNet, with common variants like VGG16 (16 layers) and VGG19 (19 layers).

Filter Sizes: A key characteristic of VGG is its use of very small (3x3) convolutional filters throughout the network. This allowed for increased depth while keeping the receptive field size similar to larger filters, but with fewer parameters and more non-linearities.

Key Features: Emphasized the importance of network depth. Demonstrated that deep networks with small, uniform filters could achieve excellent performance. Used max-pooling layers to reduce spatial dimensions.

Performance: Achieved state-of-the-art results in ILSVRC 2014. It showed that depth is a critical component for good performance, but at the cost of significantly higher computational expense and memory usage due to the large number of parameters.

```python
# 6. Using keras, build and train a simple CNN model on the MNIST dataset from scratch. Include code for module creation, compilation, training, and evaluation.

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np

# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Preprocess the data
# Reshape the images to add a channel dimension (for CNNs: (height, width, channels))
x_train = x_train.reshape(-1, 28, 28, 1).astype('float32') / 255.0
x_test = x_test.reshape(-1, 28, 28, 1).astype('float32') / 255.0

# Convert labels to one-hot encoding (for categorical crossentropy loss)
y_train = keras.utils.to_categorical(y_train, num_classes=10)
y_test = keras.utils.to_categorical(y_test, num_classes=10)

print(f"x_train shape: {x_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"x_test shape: {x_test.shape}")
print(f"y_test shape: {y_test.shape}")

# Defining the CNN model
model = keras.Sequential([
    keras.Input(shape=(28, 28, 1)),
    layers.Conv2D(32, kernel_size=(3, 3), activation='relu'),
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Conv2D(64, kernel_size=(3, 3), activation='relu'),
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Flatten(),
    layers.Dropout(0.5), # Dropout for regularization
    layers.Dense(10, activation='softmax') # 10 classes for MNIST
])

model.summary()

# Compiling the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy', # for one-hot encoded labels
              metrics=['accuracy'])

# Training the model
batch_size = 128
epochs = 10

history = model.fit(x_train, y_train,
                    batch_size=batch_size,
                    epochs=epochs,
                    validation_split=0.1) # Using 10% of training data for validation

# Evaluating the model on the test data
loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")
```

```python
# 8. Using PyTorch, write a script to define and train a CNN on the MNIST dataset. Include model definition, data loaders, training loop, and accuracy evaluation.

import torch
import torchvision
from torchvision import datasets, transforms
import numpy as np
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

# Defining transformations
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])

# Defining batch size and number of classes
batch_size = 128
num_classes = 10

# Downloading and loading the MNIST training dataset
train_dataset = datasets.MNIST('../data', train=True, download=True, transform=transform)
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

# Downloading and loading the MNIST test dataset
test_dataset = datasets.MNIST('../data', train=False, download=True, transform=transform)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

print(f"Batch size: {batch_size}")
print(f"Number of classes: {num_classes}")
print(f"Number of training samples: {len(train_dataset)}")
print(f"Number of test samples: {len(test_dataset)}")
print("MNIST dataset loaded and DataLoaders created successfully.")

# Defining the CNN Model
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        # First convolutional layer
        # 1 input channel (grayscale image), 32 output channels, 3x3 kernel
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, padding=1)
        # Second convolutional layer
        # 32 input channels, 64 output channels, 3x3 kernel
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        # Max pooling layer
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
        # Dropout layer to prevent overfitting
        self.dropout1 = nn.Dropout(0.25)
        # Fully connected layer
        # Input features are calculated based on the output of conv2 after pooling
        # For MNIST (28x28 images):
        # After conv1 (28x28), pool (14x14)
        # After conv2 (14x14), pool (7x7)
        # So, flattened size will be 64 * 7 * 7
        self.fc1 = nn.Linear(64 * 7 * 7, 128)
        # Another dropout layer
        self.dropout2 = nn.Dropout(0.5)
        # Output layer, 10 classes for MNIST
        self.fc2 = nn.Linear(128, num_classes)

    def forward(self, x):
        # Apply conv1 -> ReLU -> pool
        x = self.pool(F.relu(self.conv1(x)))
        # Apply conv2 -> ReLU -> pool
        x = self.pool(F.relu(self.conv2(x)))
        # Flatten the output for the fully connected layers
        x = x.view(-1, 64 * 7 * 7)
        # Apply dropout1
        x = self.dropout1(x)
        # Apply fc1 -> ReLU
        x = F.relu(self.fc1(x))
        # Apply dropout2
        x = self.dropout2(x)
        # Apply fc2 (output layer)
        x = self.fc2(x)
        return x

# Instantiating the model
model = Net()

# Printing the model architecture
print("CNN Model Architecture:")
print(model)

# Checking if CUDA is available and move the model to GPU if it is
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
print(f"\nModel moved to: {device}")

# Defining Loss Function and Optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

print("Loss function (CrossEntropyLoss) and Optimizer (Adam) defined successfully.")

num_epochs = 10

print(f"Starting model training for {num_epochs} epochs...")

# Training loop
for epoch in range(1, num_epochs + 1):
    model.train() # Set the model to training mode
    running_loss = 0.0
    correct_predictions = 0
    total_samples = 0
    for batch_idx, (data, target) in enumerate(train_loader):
        data, target = data.to(device), target.to(device)

        # Zero the parameter gradients
        optimizer.zero_grad()

        # Forward pass
        outputs = model(data)
        loss = criterion(outputs, target)

        # Backward pass and optimize
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * data.size(0)
        _, predicted = torch.max(outputs.data, 1)
        total_samples += target.size(0)
        correct_predictions += (predicted == target).sum().item()

    epoch_loss = running_loss / len(train_loader.dataset)
    epoch_accuracy = correct_predictions / total_samples

    print(f"Epoch {epoch}/{num_epochs}, Loss: {epoch_loss:.4f}, Accuracy: {epoch_accuracy:.4f}")

print("Model training complete.")

print("Starting model evaluation on the test set...")

model.eval() # Setting the model to evaluation mode
test_loss = 0
correct = 0
with torch.no_grad(): # Disabling gradient calculation during inference
    for data, target in test_loader:
        data, target = data.to(device), target.to(device)
        output = model(data)
        test_loss += criterion(output, target).item() * data.size(0) # Summing up batch loss
        pred = output.argmax(dim=1, keepdim=True) # Getting the index of the max log-probability
        correct += pred.eq(target.view_as(pred)).sum().item()

test_loss /= len(test_loader.dataset)
test_accuracy = correct / len(test_loader.dataset)

print(f"Final Test Loss: {test_loss:.4f}")
print(f"Final Test Accuracy: {test_accuracy:.4f}")
print("Model evaluation complete.")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Defining target image size and batch size
target_size = (150, 150)
batch_size = 32

# Creating an ImageDataGenerator instance for the training set with augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,             # Normalize pixel values to [0, 1]
    rotation_range=40,          # Rotate images by a random angle between 0 and 40 degrees
    width_shift_range=0.2,      # Shift images horizontally by a fraction of total width
    height_shift_range=0.2,     # Shift images vertically by a fraction of total height
    shear_range=0.2,            # Shear transformations
    zoom_range=0.2,             # Zoom in/out on images
    horizontal_flip=True,       # Randomly flip images horizontally
    fill_mode='nearest'         # Strategy for filling in new pixels created by transformations
)

# Creating another ImageDataGenerator instance for the validation set (only rescaling)
validation_datagen = ImageDataGenerator(rescale=1./255)

# Using flow_from_directory for training data
train_generator = train_datagen.flow_from_directory(
    'dog-cat-classifier-main/data/train',  # Directory containing training images
    target_size=target_size,               # Resize all images to target_size
    batch_size=batch_size,
    class_mode='binary',                   # Binary classification (dog/cat)
    shuffle=True                           # Shuffle training data
)

# Using flow_from_directory for validation data
validation_generator = validation_datagen.flow_from_directory(
    'dog-cat-classifier-main/data/test',   # Directory containing validation images
    target_size=target_size,
    batch_size=batch_size,
    class_mode='binary',                   # Binary classification (dog/cat)
    shuffle=False                          # Do not shuffle validation data
)

print("ImageDataGenerators and data generators (train_generator, validation_generator) created successfully.")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Initializing a Sequential model
model = Sequential()

# Adding the first convolutional layer with input shape
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(target_size[0], target_size[1], 3)))
model.add(MaxPooling2D(pool_size=(2, 2)))

# Adding the second convolutional layer
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

# Flattening the output for dense layers
model.add(Flatten())

# Adding dense layers
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='sigmoid')) # Output layer for binary classification

# Displaying the model summary
model.summary()

print("CNN model built successfully and its summary is displayed above.")
```

```python
from tensorflow.keras.optimizers import Adam

# Compiling the model
model.compile(optimizer=Adam(learning_rate=0.001),
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Defining the number of epochs
epochs = 10

# Training the model
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // batch_size,
    epochs=epochs,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // batch_size
)

print("Model compiled and trained successfully.")
```

# **Question 10:**
You are working on a web application for a medical imaging startup. Your
task is to build and deploy a CNN model that classifies chest X-ray images into “Normal”
and “Pneumonia” categories. Describe your end-to-end approach–from data preparation
and model training to deploying the model as a web app using Streamlit.

**Answer 10:**

a. Data Preparation:

Dataset Acquisition: The first step involves obtaining a suitable chest X-ray dataset. This typically consists of image files categorized into 'Normal' and 'Pneumonia' folders, often further split into 'train', 'validation', and 'test' directories.

Image Preprocessing: Before feeding images to a CNN, they need to be preprocessed. This includes:

Resizing: All images must be uniformly resized to a standard dimension to serve as consistent input for the neural network.

Normalization: Pixel values, which typically range from 0 to 255, are scaled to a smaller range (e.g., 0 to 1) by dividing by 255. This helps the model converge faster and improves performance.

Data Augmentation: To prevent overfitting and improve the model's generalization capabilities, especially with limited datasets, data augmentation techniques are applied to the training set. This involves creating modified versions of existing images by applying random transformations such as:

Shearing: Distorting the image along an axis.
Zooming: Randomly zooming in or out of the image.
Horizontal Flipping: Flipping the image horizontally.
Rotation, Brightness changes, etc. can also be included.

Dataset Splitting: The dataset is explicitly divided into three sets:
Training Set: Used to train the model, including data augmentation.
Validation Set: Used during training to monitor the model's performance on unseen data and to tune hyperparameters, without data augmentation.
Test Set: A completely independent set of images used only after training to evaluate the final model's performance, also without data augmentation.

Data Generators: For efficiency, especially with large image datasets, ImageDataGenerator (from Keras) is used to load images in batches directly from directories, apply preprocessing and augmentation on-the-fly, and prepare them for model training.

b. Build and Train CNN Model:

Model Architecture Definition: A Convolutional Neural Network (CNN) is designed. A typical architecture includes:

Convolutional Layers (Conv2D): These layers apply filters to input images, detecting features like edges, textures, and patterns. They learn hierarchical representations of the data.
Activation Functions: Commonly ReLU (Rectified Linear Unit) is used after convolutional layers to introduce non-linearity.
Pooling Layers (MaxPooling2D): These layers reduce the spatial dimensions of the feature maps, reducing computational complexity and helping to extract dominant features.
Flatten Layer: After several convolutional and pooling layers, the 2D feature maps are flattened into a 1D vector.
Dense (Fully Connected) Layers: These layers perform classification based on the features extracted by the convolutional layers.
Dropout Layers: Introduced between dense layers to randomly deactivate a fraction of neurons during training. This regularization technique helps prevent overfitting.
Output Layer: For binary classification (Normal/Pneumonia), a single dense layer with a 'sigmoid' activation function is used, outputting a probability score between 0 and 1.
Model Compilation: Before training, the model must be compiled with:

Accuracy: The proportion of correctly classified instances.
Precision: The proportion of true positive predictions among all positive predictions.
Recall (Sensitivity): The proportion of true positive predictions among all actual positive instances.
